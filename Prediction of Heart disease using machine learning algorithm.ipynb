# Machine Learning

## Data Pre-Processing

```python
columns = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"]
data = pd.read_csv("C:/Users/dixit/Desktop/heart disease/heart-disease.csv")
print(data.head())
```

```python
from sklearn.preprocessing import StandardScaler
X = data.drop('target', axis=1)
y = data['target']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

# K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)  
y_pred = knn.predict(X_test)

# Evaluate the model
print("KNN Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
cm_knn = confusion_matrix(y_test, y_pred)

# Plot Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='magma')
plt.title('k-NN Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
```

```python
def plot_roc_curve(fpr, tpr, model_name):
    plt.plot(fpr, tpr, color='red', label=f'{model_name} (AUC = {auc(fpr, tpr):.2f})')
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred_prob_knn = knn.predict_proba(X_test)[:, 1]

fpr_knn, tpr_knn, _ = roc_curve(y_test, y_pred_prob_knn)
plot_roc_curve(fpr_knn, tpr_knn, "k-NN")
```

# Comparison among the ML models

```python
# Define the classifiers to be evaluated
Classifiers = [
    LogisticRegression(max_iter=1000, random_state=42),
    KNeighborsClassifier(),
    DecisionTreeClassifier(random_state=42),
    RandomForestClassifier(random_state=42),
    AdaBoostClassifier(random_state=42),
    SVC(kernel='linear', random_state=42),
    GaussianNB(),
    MLPClassifier(max_iter=1000, random_state=42),
    lgb.LGBMClassifier(random_state=42),
    xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
]

# Initialize lists to store accuracy and model names
Accuracy = []
Model = []

# Iterate over the classifiers and evaluate them
for classifier in Classifiers:
    classifier.fit(X_train, y_train)
    pred = classifier.predict(X_test)
    accuracy = accuracy_score(y_test, pred)
    Accuracy.append(accuracy)
    Model.append(classifier.__class__.__name__)
    print(f'Accuracy of {classifier.__class__.__name__} is {accuracy}')

# Create a DataFrame to store the results
results_df = pd.DataFrame({'Model': Model, 'Accuracy': Accuracy})

# Sort the results by accuracy
results_df = results_df.sort_values(by='Accuracy', ascending=False)

# Plot the results with dark blue color as vertical bars
plt.figure(figsize=(12, 8))
plt.bar(results_df['Model'], results_df['Accuracy'], color='Lightblue')
plt.ylabel('Accuracy')
plt.xlabel('Model')
plt.title('Comparison of ML Model Accuracies on UCI Heart Disease Dataset')
plt.xticks(rotation=45, ha="right")
plt.show()
```
